Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
res['income_binary'] = res['income'].map({'<=50K': 0, '>50K': 1})
display(res[['income', 'income_binary']].head())
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
bins = [0, 30, 60, res['age'].max() + 1]
labels = ['Youth', 'Adult', 'Senior']
res['age_group'] = pd.cut(res['age'], bins=bins, labels=labels, right=False)
display(res[['age', 'age_group']].head())
```

Rename the 'educational-num' column to 'education_num'.

```python
res.rename(columns={'educational-num': 'education_num'}, inplace=True)
display(res.columns)
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
res['workclass'] = res['workclass'].replace('?', 'Unknown')
res['occupation'] = res['occupation'].replace('?', 'Unknown')

print("Unique values in 'workclass' after replacement:")
display(res['workclass'].unique())

print("Unique values in 'occupation' after replacement:")
display(res['occupation'].unique())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = res.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = res.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

Get the count of each marital-status category.

```python
marital_status_counts = res['marital-status'].value_counts()
display(marital_status_counts)
```

Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_by_education = res.groupby('education')['hours-per-week'].mean()
display(mean_hours_by_education)
```

Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = res.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

Get the unique values and their counts for the 'native-country' column

```python
native_country_counts = res['native-country'].value_counts()
display(native_country_counts)
```

Filter the dataset to include only individuals with income >50K.

```python
income_greater_50k = res[res['income'] == '>50K']
display(income_greater_50k.head())
```

Sort the dataset by age in descending order.

```python
res_sorted_by_age = res.sort_values(by='age', ascending=False)
display(res_sorted_by_age.head())
```

Filter the dataset to include only individuals who work more than 40 hours per week.

```python
hours_greater_than_40 = res[res['hours-per-week'] > 40]
display(hours_greater_than_40.head())
```

Calculate the median income by education level and gender.


```python
median_income_by_education_gender = res.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

Calculate the median income by education level and gender.

```python
res['income_binary'] = np.where(res['income'] == '<=50K', 0, 1)
median_income_by_education_gender = res.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

Find the top 3 occupations with the highest average capital-gain.

```python
top_occupations_capital_gain = res.groupby('occupation')['capital-gain'].mean().sort_values(ascending=False).head(3)
display(top_occupations_capital_gain)
```

Filter the dataset to include individuals who are 'Married-civ-spouse' and have an income >50K.

```python
married_high_income = res[(res['marital-status'] == 'Married-civ-spouse') & (res['income'] == '>50K')]
display(married_high_income.head())
```

Create 5 different charts and write your observations from the chart as comment

```python
plt.figure(figsize=(10, 6))
sns.histplot(res['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=res, x='gender', hue='income', palette='viridis')
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Income')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=res, y='workclass', palette='tab10')
plt.title('Frequency of Different Workclass Categories')
plt.xlabel('Count')
plt.ylabel('Workclass')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=res, y='workclass', hue='workclass', palette='tab10', legend=False)
plt.title('Frequency of Different Workclass Categories')
plt.xlabel('Count')
plt.ylabel('Workclass')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=res.groupby('education')['income_binary'].mean().reset_index(),
            x='education', y='income_binary', palette='viridis')
plt.title('Average Income (Binary) by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Income (>50K = 1, <=50K = 0)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
