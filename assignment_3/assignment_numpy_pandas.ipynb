4. Generate a 1D array of 15 random integers between 1 and 100 and find its min, max, mean, and standard deviation.

```python
import numpy as np

arr = np.random.randint(1, 101, size=15)

minimum = np.min(arr)
maximum = np.max(arr)
mean = np.mean(arr)
std_dev = np.std(arr)


print("Array:", arr)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("Mean:", mean)
print("Standard Deviation:", std_dev)
```

5. Given two arrays of sales and costs, calculate profit and profit percentage arrays.

```python
import numpy as np


sales = np.array([1000, 2500, 800, 3200, 1500, 900])
costs = np.array([ 700, 1800, 600, 2100, 1200, 950])
profit = sales - costs
profit_percentage = (profit / costs) * 100
profit_percentage = np.round(profit_percentage, 2)
print("Sales:            ", sales)
print("Costs:            ", costs)
print("Profit:           ", profit)
print("Profit %:         ", profit_percentage, "%")
```

6. Create a 6×6 random matrix and compute row sums, column sums, and total sum.

```python
import numpy as np


np.random.seed(42)

matrix = np.random.randint(1, 101, size=(6, 6))

print("6×6 Random Matrix:")
print(matrix)
print()

row_sums = matrix.sum(axis=1)

col_sums = matrix.sum(axis=0)


total_sum = matrix.sum()


print("Row sums:     ", row_sums)
print("Column sums:  ", col_sums)
print("Total sum:    ", total_sum)
```

7. Given stock prices, compute daily returns using vectorized NumPy operations.

```python
import numpy as np

prices = np.array([100.0, 102.5, 101.0, 105.0, 107.5, 106.0, 110.0, 108.5, 112.0, 115.0])

daily_returns = prices[1:] / prices[:-1] - 1



print("Stock Prices:     ", prices)
print("Daily Returns:    ", np.round(daily_returns, 4))
```

8. Convert an array of annual interest rates into monthly rates using vectorization.

```python
import numpy as np

annual_rates = np.array([0.03, 0.05, 0.06, 0.08, 0.12, 0.00])


monthly_rates = (1 + annual_rates) ** (1/12) - 1

monthly_rates_pct = np.round(monthly_rates * 100, 4)


print("Annual Rates (%):     ", annual_rates * 100)
print("Monthly Rates (%):    ", monthly_rates_pct)
```

9. Normalize a given array of portfolio weights so that the sum becomes 1.

```python
import numpy as np

raw_weights = np.array([30, 25, 20, 15, 10, 5])


normalized_weights = raw_weights / raw_weights.sum()


normalized_weights = np.round(normalized_weights, 4)

print("Raw weights:         ", raw_weights)
print("Sum of raw weights:  ", raw_weights.sum())
print("Normalized weights:  ", normalized_weights)
print("Sum after normalization:", normalized_weights.sum())
```

10. Create a Pandas Series from the following list: [1200, 1500, 1800, 2000, 2200]

a. Find:

i. mean

ii. max

iii. index of the max value

```python
import pandas as pd


data = [1200, 1500, 1800, 2000, 2200]
sales = pd.Series(data)

print(sales)

# i. Mean
mean_value = sales.mean()

# ii. Maximum value
max_value = sales.max()

# iii. Index of the maximum value
max_index = sales.idxmax()

print("i.   Mean                 :", mean_value)
print("ii.  Maximum value        :", max_value)
print("iii. Index of max value   :", max_index)
```

11. Create a DataFrame with Student Name + 3 subject marks:

Name | Math | Science | English

Then:

· Print only the Math column

· Print first 2 rows

· Print the shape

```python
import pandas as pd

# Create the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math': [85, 92, 78, 96, 88],
    'Science': [90, 88, 95, 91, 87],
    'English': [82, 85, 89, 94, 90]
}

df = pd.DataFrame(data)

# Display the full DataFrame
print("Full DataFrame:")
print(df)
print()

# Tasks:
# 1. Print only the Math column
print("Math column only:")
print(df['Math'])
# Alternative: print(df.Math)

print("\nMath column (as DataFrame):")
print(df[['Math']])   # double brackets → keeps it as DataFrame

# 2. Print first 2 rows
print("\nFirst 2 rows:")
print(df.head(2))

# 3. Print the shape
print("\nShape of DataFrame:", df.shape)
print("→", df.shape[0], "rows and", df.shape[1], "columns")
```

12. Using this DataFrame:

df = pd.DataFrame({

"Account": [101,102,103,104],

"Balance": [5000, 15000, 7000, 12000]

})

Filter rows where:

· Balance > 8000

· Balance between 6000 and 13000

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({
    "Account": [101, 102, 103, 104],
    "Balance": [5000, 15000, 7000, 12000]
})

print("Original DataFrame:")
print(df)
print()
high_balance = df[df['Balance'] > 8000]
print("Accounts with Balance > 8000:")
print(high_balance)
mid_balance = df[df['Balance'].between(6000, 13000)]
# or: df[(df['Balance'] >= 6000) & (df['Balance'] <= 13000)]

print("\nAccounts with Balance between 6000 and 13000:")
print(mid_balance)
```
