```python
df = pd.read_csv("/content/Bengaluru_House_Data.csv")
```

```python
print(df['society'].isnull().sum())
```

```python
df['society'] = df['society'].fillna('No Society')
print(df['society'].isnull().sum())
display(df.head())
```

```python

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]

print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")

print(f"\nNumber of outliers identified: {len(outliers)}")
print("\nOutliers in the 'price' column:")
display(outliers[['price', 'location', 'total_sqft']])
```

Skews Statistics: Extreme prices distort the mean and can mislead interpretations of central tendency and distribution.
Harms Model Performance: Outliers can degrade machine learning models (e.g., linear regression) by pulling predictions toward rare cases, reducing accuracy on typical data.
Domain Relevance: In real estate, these extremes often represent unique or anomalous properties (e.g., luxury estates or distressed sales) that don't reflect the general market.
